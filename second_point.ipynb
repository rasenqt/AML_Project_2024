```python
import open3d
import numpy as np

pcd = open3d.io.read_point_cloud('./8th_scan_angel.ply')
pcd.estimate_normals(search_param=open3d.geometry.KDTreeSearchParamHybrid(radius=0.1, max_nn=30))

with open3d.utility.VerbosityContextManager(open3d.utility.VerbosityLevel.Debug) as cm:
    mesh, densities = open3d.geometry.TriangleMesh.create_from_point_cloud_poisson(pcd, depth=9)
    open3d.io.write_triangle_mesh('./angel.obj', mesh, write_ascii=False, compressed=False, write_vertex_normals=True, write_vertex_colors=True, write_triangle_uvs=True, print_progress=False)
```
